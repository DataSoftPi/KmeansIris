In this project, we will explore and learn about the K-Means clustering algorithm. We will start with simple synthetic data and gradually progress to real-world datasets to understand the concepts, implementation, and evaluation of K-Means clustering.

Synthetic Data Exploration
We will begin by generating synthetic data with clear separation for different clusters. This dataset will have 100 elements, 10 dimensions, and 5 different clusters. To visualize the data, we will use t-SNE (t-Distributed Stochastic Neighbor Embedding) to project the high-dimensional data into a 2D space.

We will implement our own K-Means clustering algorithm and evaluate its performance against the ground truth clustering. Additionally, we will compare it with the K-Medians algorithm, which uses medians instead of means to estimate the cluster centers.

```python
# Import necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import adjusted_mutual_info_score
from sklearn.manifold import TSNE, MDS
```

```python
# Synthetic Data Exploration

# Generate synthetic data
X, y = datasets.make_blobs(n_samples=100, centers=5, n_features=10, random_state=42, cluster_std=2, center_box=(0, 10))

# Visualize synthetic data using t-SNE
dims = TSNE(random_state=42).fit_transform(X)
dim1, dim2 = dims[:, 0], dims[:, 1]
sns.scatterplot(x=dim1, y=dim2, hue=y, palette='tab10', legend=False)
plt.title('Data Distribution after t-SNE ')
plt.show()
```

Here we created some blobs and used TSNE to reduce the dimension which helps to make the viusalization easier

```python
# Compare K-Means clustering results with ground truth
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)
y_pred_kmeans = kmeans.labels_
sns.scatterplot(x=dim1, y=dim2, hue=y_pred_kmeans, palette='tab10', legend=False)
plt.title('K-Means Clustering Results')
plt.show()
```


In this section, we will observe the clustering progress of the K-Means algorithm by iterating the number of iterations.
We will visualize the clustering results after each iteration to see how the clusters evolve.


```python
fig, axes = plt.subplots(3, 1, figsize=(7.5, 8), sharey=True)
fig.tight_layout()
plt.subplots_adjust(hspace=0.3)
for i in range(3):
    kmeans = KMeans(n_clusters=5, max_iter=i+1, random_state=42)
    kmeans.fit(X)
    y_pred_kmeans = kmeans.labels_
    ax = axes[i]
    ax.title.set_text(f'K-Means in Iteration {i}')
    sns.scatterplot(x=dim1, y=dim2, hue=y_pred_kmeans, palette='tab10', ax=ax, legend=False)
plt.show()
```

In this section, we will evaluate the clustering results by comparing them to the ground truth labels.
We will use the adjusted mutual information score to measure the agreement between the predicted labels and the true labels.


```python
score = adjusted_mutual_info_score(y_pred_kmeans, y)
print(f"Mutual Info Score: {score}")
```

In this section, we will explore real-world data using the famous iris dataset from sklearn.datasets module.
The dataset consists of 150 samples with 4 features and is divided into 3 classes (species of iris flowers).
We will visualize the data using MDS to obtain a 2D representation of the data distribution.


```python
# Load the iris dataset
data = datasets.load_iris()
X, y = data['data'], data['target']

# Visualize real-world data using MDS
dims = MDS(random_state=42).fit_transform(X)
dim1, dim2 = dims[:, 0], dims[:, 1]
sns.scatterplot(x=dim1, y=dim2, hue=y, palette='tab10', legend=False)
plt.title('Data Distribution after MDS (Real-World Data)')
plt.show()
```

```python
# Compare K-Means clustering results with ground truth
kmeans = KMeans( random_state=42)
kmeans.fit(X)
y_pred_kmeans = kmeans.labels_
sns.scatterplot(x=dim1, y=dim2, hue=y_pred_kmeans, palette='tab10', legend=False)
plt.title('K-Means Clustering Results')
plt.show()
```

```python
# Iterative Clustering Progress


fig, axes = plt.subplots(6, 1, figsize=(7.5, 16), sharey=True)
fig.tight_layout()
plt.subplots_adjust(hspace=0.3)
for i in range(6):
    kmeans = KMeans( max_iter=i+1, random_state=42)
    kmeans.fit(X)
    y_pred_kmeans = kmeans.labels_
    ax = axes[i]
    ax.title.set_text(f'K-Means in Iteration {i}')
    sns.scatterplot(x=dim1, y=dim2, hue=y_pred_kmeans, palette='tab10', ax=ax, legend=False)
plt.show()
```

```python
score = adjusted_mutual_info_score(y_pred_kmeans, y)
print(f"Mutual Info Score: {score}")
```

The results are not very good since it is supposed to have 3 different species. ELbow method is a very handy method to check how many clusters are there

```python
# Elbow Method for Optimal Number of Clusters
"""
In this section, we will use the Elbow method to determine the optimal number of clusters (k) for the iris dataset.
We will calculate the inertia (within-cluster sum of squares) for different values of k and plot it.
"""

inertias = []
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot the Elbow curve
plt.plot(k_values, inertias, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method: Inertia vs. Number of Clusters')
plt.show()
```

as the name suggest elbow method is used to pick number of clusters in the figure above. we can see that at 3 there is a elbow show that should be the number of clusters

```python
kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(X)
y_pred_kmeans = kmeans.labels_
sns.scatterplot(x=dim1, y=dim2, hue=y_pred_kmeans, palette='tab10', legend=False)
plt.title('K-Means Clustering Results')
plt.show()
```

```python
score = adjusted_mutual_info_score(y_pred_kmeans, y)
print(f"Mutual Info Score: {score}")
```

Lets try using other algorithms available in kmeans to check if they perform better

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'init': ['k-means++', 'random'],
    'max_iter': [100, 200, 300]
}

# Create a KMeans instance
kmeans = KMeans(n_clusters=3, random_state=42)

# Perform Grid Search
grid_search = GridSearchCV(kmeans, param_grid, cv=5)
grid_search.fit(X)


```

```python
# Print the best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
#lets create the model with these parameters
kmeans = KMeans(n_clusters=3, random_state=42,init='k-means++', max_iter=100)
kmeans.fit(X)
y_pred_kmeans = kmeans.labels_
sns.scatterplot(x=dim1, y=dim2, hue=y_pred_kmeans, palette='tab10', legend=False)
plt.title('K-Means Clustering Results')
plt.show()
```

```python
score = adjusted_mutual_info_score(y_pred_kmeans, y)
print(f"Mutual Info Score: {score}")
```

From the K-means experiments and analysis conducted, we can draw several key insights and learnings:

Data Clustering: K-means clustering is a popular unsupervised machine learning algorithm used for data clustering. It aims to group similar data points together based on their features and minimize the within-cluster sum of squares.

Synthetic Data Experiment: We started by applying K-means on synthetic data with clear separation between clusters. The algorithm performed well, accurately identifying the clusters. The choice between K-means and K-medians depends on the presence of outliers and the underlying data distribution.

Real-World Data Experiment: We then applied K-means to real-world data using the iris dataset. The algorithm struggled to accurately cluster the data, indicating the limitations of K-means on complex datasets with overlapping classes. The choice of the number of clusters (k) is crucial and can be determined using techniques like the Elbow method.

Elbow Method: The Elbow method helps in selecting the optimal number of clusters by analyzing the inertia (within-cluster sum of squares) for different values of k. The elbow point on the inertia plot suggests the number of clusters at which further division does not provide significant improvement.

Hyperparameter Tuning: We performed hyperparameter tuning for K-means by using Grid Search to search for the best combination of hyperparameters. We focused on tuning hyperparameters like the initialization method and maximum number of iterations.

Insights and Limitations: Through the experiments, we gained insights into the strengths and limitations of the K-means algorithm. K-means can work well on simple datasets with well-separated clusters but may struggle with complex datasets and overlapping classes. The performance can be influenced by the choice of initialization, presence of outliers, and data distribution.

Further Improvements: To improve K-means performance, one can explore advanced initialization methods like k-means++, use different distance metrics, or consider alternative clustering algorithms like Gaussian Mixture Models (GMM) or Density-Based Spatial Clustering of Applications with Noise (DBSCAN) for more complex datasets.
